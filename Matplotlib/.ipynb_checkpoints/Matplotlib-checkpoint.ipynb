```python
import numpy as np
import matplotlib as mp
import matplotlib.pyplot as plt
```

### 1.1 Generating Sin and Cos Graph

```python
x = np.linspace(0,3* np.pi, 100)

sin_y = np.sin(x)
cos_y = np.cos(x)

plt.plot(x, sin_y, label = 'Sine')
plt.plot(x, cos_y, label = 'Cosine')

plt.legend(loc = 'upper right')

plt.title('Wave')
plt.show()
```

```python
x = np.linspace(0,3* np.pi, 100)

sin_y = np.sin(x)
cos_y = np.cos(x)

plt.plot(x, sin_y, '*')
plt.plot(x, cos_y, '--')

plt.title('Wave')
plt.show()
```

#### 4.1 Line Plot

```python
x = np.linspace(0,3* np.pi, 100)
y = np.sin(x)

plt.plot(x,y)

plt.title('Sine Wave')
plt.show()
```

#### 4.1.2 Multiple Line Plots

```python
x = np.linspace(0,3* np.pi, 100)
y_sin = np.sin(x)
y_cos = np.cos(x)

plt.figure(figsize = (15,5))

plt.plot(x,y_sin, '*', color = 'red', linewidth = 4)
plt.plot(x,y_cos, '--', color = 'blue', linewidth = 3)

plt.title('Sine Wave')
plt.show()
```

### 4.1.3 Line Styles

```python
plt.figure(figsize = (15,5))

plt.plot(x, x+0, linestyle = 'solid')
plt.plot(x, x+1, linestyle = 'dotted')
plt.plot(x, x+2, linestyle = 'dashed')
plt.plot(x, x+3, linestyle = 'dashdot')

plt.show()
```

### 4.2) Scatter PLot

```python
x = np.random.rand(300)
y = np.random.rand(300)

plt.scatter(x,y, c='blue', alpha = .5, s=40)
plt.title('Scatter Plot')
plt.show()
```

#### 4.2.2 Multiple Scatter Plots

```python
n = 500
x1 = np.random.rand(n)
y1 = np.random.rand(n)

x2 = np.random.rand(n) + 2
y2 = np.random.rand(n) + 2

plt.scatter(x1,y1, c='blue', alpha = .5, s=20, label = 'Group 1')
plt.scatter(x2,y2, c='red', alpha = .5, s=20, label = 'Group 2')
plt.title('Scatter Plot with Multiple Groups')
plt.legend()
plt.show()
```

#### 4.2.3 On Real World Dataset

```python
from sklearn.datasets import load_iris

iris = load_iris()
x = iris.data
y = iris.target

plt.scatter(x[:,0] , x[:,1])
```

```python
plt.scatter(x[:,0] , x[:,1], c='blue', alpha = 0.8, marker = '+')
```

```python
rng = np.random.RandomState(0)

x = rng.randn(100)
y = rng.randn(100)

colors = rng.rand(100)
sizes = rng.rand(100) * 500

plt.scatter(x,y, c = colors , s= sizes, alpha = .8, cmap = 'viridis', edgecolors = 'black')
plt.show()
```

## 4.3) Histograms and Binnings

```python
data = np.random.randn(1000)

plt.hist(data)

plt.show()
```

```python
data = np.random.randn(1000)

plt.hist(data, bins = 30, edgecolor = 'black')

plt.xlabel('Values')
plt.ylabel('Frequency')

plt.show()
```

```python
x1 = np.random.randn(1000)
x2 = np.random.randn(1000) + 2
x3 = np.random.randn(1000) + 3

plt.hist(x1, bins = 30, alpha = .8)
plt.hist(x2, bins = 30, alpha = .8)
plt.hist(x3, bins = 30, alpha = .8)

plt.xlabel('Values')
plt.ylabel('Frequency')

plt.show()
```

#### Top View of Histogram

```python
x = np.random.randn(1000)
y = np.random.randn(1000) + 2

plt.hist2d(x,y, bins = 30)

plt.xlabel('x')
plt.ylabel('y')
plt.title('2D Histogram')

plt.show()
```

## 4.5) Pie Charts

```python
cars = ['AUDI', 'BMW' , 'FORD' , 'TESLA' , 'JAGUAR' , 'MERCEDES']

data = [450 , 385 , 40 , 15, 200 , 500]

fig = plt.figure(figsize = (7,7))

plt.pie(data, labels = cars)
plt.show()
```

```python
cars = ['AUDI', 'BMW' , 'FORD' , 'TESLA' , 'JAGUAR' , 'MERCEDES']

data = [450 , 385 , 40 , 15, 200 , 500]
explode = [0, 0, .1, 0, 0, 0]

fig = plt.figure(figsize = (7,7))

plt.pie(data, labels = cars, explode = explode)
plt.show()
```

```python
# Adding Shadows
cars = ['AUDI', 'BMW' , 'FORD' , 'TESLA' , 'JAGUAR' , 'MERCEDES']

data = [450 , 385 , 40 , 15, 200 , 500]
explode = [0, 0, .1, 0, 0, 0]

fig = plt.figure(figsize = (7,7))

plt.pie(data, labels = cars, explode = explode, shadow = True)
plt.show()
```

```python
## Adding Percentage

cars = ['AUDI', 'BMW' , 'FORD' , 'TESLA' , 'JAGUAR' , 'MERCEDES']

data = [450 , 385 , 40 , 15, 200 , 500]
explode = [0, 0, .1, 0, 0, 0]

fig = plt.figure(figsize = (7,7))

plt.pie(data, labels = cars, explode = explode, shadow = True, autopct = '%1.2f%%')
plt.show()
```
